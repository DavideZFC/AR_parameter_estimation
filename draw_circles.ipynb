```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_circular_crown(r1, r2, n, center_x, center_y, ax, color='C0'):
    # Create the grid of points
    x, y = np.meshgrid(np.linspace(center_x-r2, center_x+r2, n), np.linspace(center_y-r2, center_y+r2, n))

    # Compute the distance of each point from the origin
    r = np.sqrt(x**2 + y**2)

    # Filter out any points that lie outside the circular crown
    x = x[(r >= r1) & (r <= r2)]
    y = y[(r >= r1) & (r <= r2)]

    # Plot the grid of points in the circular crown
    ax.scatter(x, y, s=2, color=color)

    # Set the plot limits and axis labels
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    ax.set_xlabel('x')
    ax.set_ylabel('y')

    # Add the two circles that define the circular crown to the plot
    circle1 = plt.Circle((0, 0), r1, color='r', fill=False)
    circle2 = plt.Circle((0, 0), r2, color='r', fill=False)
    ax.add_artist(circle1)
    ax.add_artist(circle2)
```

```python


# Show the plot
r = 0.9
center_x = 0
center_y = 0
n1 = 10
N = 5
n0 = int(n*N**0.5)
fig, ax = plt.subplots(1,2)


plot_circular_crown(0, r, n0, center_x, center_y, ax[0])
ax[0].scatter(x=center_x, y=center_y, color='C1', marker='x', label='True parameter')
ax[0].set_aspect(1)
ax[0].legend()

for i in range(N):
    plot_circular_crown(r/2, r, n1, center_x, center_y, ax[1])
    r /= 2

ax[1].set_aspect(1)
ax[1].scatter(x=center_x, y=center_y, color='C1', marker='x', label='True parameter')
ax[1].legend()
fig.subplots_adjust(wspace=0.3)
fig.savefig('figures/circles.pdf')
fig.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

# Plot the data on each subplot
ax1.plot(x, y1)
ax2.plot(x, y2)

# Set the labels for each subplot
ax1.set_ylabel('y1')
ax2.set_ylabel('y2')
ax2.set_xlabel('x')

# Adjust the spacing between the subplots
fig.subplots_adjust(vspace=10.5)

# Show the plot
plt.show()
```
